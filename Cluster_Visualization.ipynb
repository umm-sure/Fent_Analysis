```python
import pandas as pd
df_unnormalized = pd.read_csv("national_data_with_null_unnormalized.csv")
df_clusters = pd.read_csv("nationalgyat.csv")
```

```python
df_unnormalized
```

```python
del df_unnormalized['Number of Pharmacies']
```

```python
df_clusters = df_clusters[['County Name', 'cluster']]
```

```python
# prompt: df_unnormalized join on County Name with df_clusters


df_unnormalized=df_unnormalized.merge(df_clusters, how='left', on=['County Name'])
```

```python
df_unnormalized.dropna()
```

```python
del df_unnormalized['Unnamed: 0']
```

```python
# prompt: get the average value for each column for each cluster

df_unnormalized.groupby(['cluster']).mean().to_csv("cluster_attribs.csv")
```

```python
df_unnormalized.groupby(['cluster']).mean()
```

```python
# prompt: replace combined_df with df in this code  = combined_df['Hospital_Norm'] = combined_df['Hospitals'] / combined_df['Population'] combined_df['Traffic_Norm'] = combined_df['Traffic_Volume'] / combined_df['Population'] combined_df['Enforcement_Norm'] = combined_df['Number of Local Law Enforcements'] / combined_df['Population'] combined_df['Fire_Norm'] = combined_df['Number of Fire Departments'] / combined_df['Population']  combined_df['Hospital_Norm'] = combined_df['Hospitals'] / combined_df['Hospitals'].max() combined_df['Traffic_Norm'] = combined_df['Traffic_Volume'] / combined_df['Traffic_Volume'].max() combined_df['Enforcement_Norm'] = combined_df['Number of Local Law Enforcements'] / combined_df['Number of Local Law Enforcements'].max() combined_df['Fire_Norm'] = combined_df['Number of Fire Departments'] / combined_df['Number of Fire Departments'].max()
df = df_unnormalized.copy()
df['Hospitals'] = df['Hospitals'] / df['Population'] * 10000000
df['Traffic_Volume'] = df['Traffic_Volume'] / df['Population'] * 100
df['Number of Local Law Enforcements'] = df['Number of Local Law Enforcements'] / df['Population'] * 10000000
df['Number of Fire Departments'] = df['Number of Fire Departments'] / df['Population'] * 10000000

df['Number of Roads'] = df['Number of Roads'] / df['Area_Sq_Miles']
df['Number of Roads'] = df['Number of Roads'] / df['Number of Roads'].max() * 100

df['Overdoses_Norm'] = df['Overdoses'] /  df['Population'] *100000
```

```python
df['Population_Density_Norm'] = df['Population_Density'] / df['Population_Density'].max() *20
df['Population_Norm'] = df['Population'] / df['Population'].max() * 20
```

```python
df.groupby(['cluster']).mean().to_csv("cluster_attribs_normalized.csv")
```

```python
df.groupby(['cluster']).mean()
```

```python
# make all of these columns between 0 and 1 df[['Hospitals', 'Shannon_Weiner', 'Traffic_Volume', 'Number of Local Law Enforcements', 'Number of Fire Departments', 'Number of Roads', 'Population_Norm','Population_Density_Norm']]

df[['Hospitals', 'Shannon_Weiner', 'Traffic_Volume', 'Number of Local Law Enforcements', 'Number of Fire Departments', 'Number of Roads', 'Population_Norm','Population_Density_Norm']] = df[['Hospitals', 'Shannon_Weiner', 'Traffic_Volume', 'Number of Local Law Enforcements', 'Number of Fire Departments', 'Number of Roads', 'Population_Norm','Population_Density_Norm']].apply(lambda x: (x-x.min()) / (x.max() - x.min()))
```

```python
# make all of these columns between 0 and 1 df[['Hospitals', 'Shannon_Weiner', 'Traffic_Volume', 'Number of Local Law Enforcements', 'Number of Fire Departments', 'Number of Roads', 'Population_Norm','Population_Density_Norm']]

df[['Hospitals', 'Shannon_Weiner', 'Traffic_Volume', 'Number of Local Law Enforcements', 'Number of Fire Departments', 'Number of Roads', 'Population_Norm','Population_Density_Norm']] = df[['Hospitals', 'Shannon_Weiner', 'Traffic_Volume', 'Number of Local Law Enforcements', 'Number of Fire Departments', 'Number of Roads', 'Population_Norm','Population_Density_Norm']].apply(lambda x: (x-x.min()) / (x.max() - x.min()))
```

```python
#  visualize a bar plot comparing each feature

import matplotlib.pyplot as plt
df = df.groupby(['cluster']).mean()
df_plot = df[['Hospitals', 'Shannon_Weiner', 'Traffic_Volume', 'Number of Local Law Enforcements', 'Number of Fire Departments', 'Number of Roads', 'Population_Norm','Population_Density_Norm']]
df_plot.plot(kind='bar', stacked=True)
plt.xlabel('Cluster')
plt.ylabel('Normalized Value')
plt.title('Comparison of Clusters')
plt.show()
```
